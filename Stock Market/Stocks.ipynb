```python
import pandas as pd
import matplotlib.pyplot as plt


JPM = pd.read_csv("JPM.csv", sep=",")       #JPMorgan Chase
BOC= pd.read_csv("3988.HK.csv", sep=",")    #Bank of China
BAC = pd.read_csv("BAC.csv", sep=",")       #Bank of America
C = pd.read_csv("C.csv", sep=",")           #Citigroup
CS = pd.read_csv("CS.csv", sep=",")         #Credit Suisse
DB = pd.read_csv("JPM.csv", sep=",")        #Deutchse Bank
HSBC = pd.read_csv("HSBC.csv", sep=",")     #HSBC
SMBC = pd.read_csv("SMBC.csv", sep=",")     #SMBC group
WFC = pd.read_csv("WFC.csv", sep=",")       #Wells Fargo


ax = WFC.plot(x= "Date", y="Open", figsize=[20,13], label="WFC", title="Stock Market Jan 1st - March 24th")
SMBC.plot(x = 'Date', y = 'Open', ax=ax, label="SMBC")
HSBC.plot(x = 'Date', y = 'Open', ax=ax, label="HSBC")
DB.plot(x = 'Date', y = 'Open', ax=ax, label="DB")
CS.plot(x = 'Date', y = 'Open', ax=ax, label="CS")
C.plot(x = 'Date', y = 'Open', ax=ax, label="C")
BAC.plot(x = 'Date', y = 'Open', ax=ax, label="BAC")
BOC.plot(x = 'Date', y = 'Open', ax=ax, label="BOC")
JPM.plot(x = 'Date', y = 'Open', ax=ax, label="JPM")
ax.set_ylabel("Open market price", size=16)
plt.show()

```
